```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

np.random.seed(30)
n = 70 # number of sample to be drawn
mu = [mu0, mu1, mu2]
sigma = [sigma0, sigma1, sigma2]
samples = []

for i in range(n): # iteratively draw samples
    Z = np.random.choice([0,1,2]) # latent variable
    samples.append(np.random.normal(mu[Z], sigma[Z], 1))

sns.distplot(samples, hist=True)
plt.xlim(0,180)
plt.title("Mixed distribution of three consecutive bars")
plt.savefig("imgs/mixturedist2.png")
plt.show()
```
